```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.naive_bayes import GaussianNB
from mlxtend.plotting import plot_decision_regions
```

```python
#read data
data = pd.read_csv('binclass.txt', sep = ",", header = None, names = ['X1', 'X2', 'y'])
```

```python
#X is two columns and y is one column
X1 = data['X1']
X2 = data['X2']
y  = data['y']
```

```python
plt.scatter(X1[y == -1], X2[y == -1])
plt.scatter(X1[y == 1], X2[y == 1])
plt.show()
```

```python
X = np.array(data.iloc[:, 0:2])
y = np.array(data.iloc[:, 2])

model = GaussianNB()
model.fit(X, y)
Yhat = model.predict(X)
print(f'Accuracy = {round(np.mean(y == Yhat) * 100, 2)} %')
```

```python
x1_min, x1_max = X1.min() - 0.1, X1.max() + 0.1
x2_min, x2_max = X2.min() - 0.1, X2.max() + 0.1
x1 = np.linspace(x1_min, x1_max, 100)
x2 = np.linspace(x2_min, x2_max, 100)
x1, x2 = np.meshgrid(x1, x2)
X = np.vstack((np.ravel(x1), np.ravel(x2))).T
Yhat = model.predict(X)

plt.contourf(x1, x2, Yhat.reshape(x1.shape), cmap=plt.cm.RdYlBu, alpha = 0.7)
plt.scatter(X1[y == -1], X2[y == -1])
plt.scatter(X1[y == 1], X2[y == 1])
plt.show()
```
